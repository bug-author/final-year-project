```python
def build_model():
    model = tensorflow.keras.models.Sequential()
    model.add(tensorflow.keras.layers.LSTM(18, input_shape=(4,1),return_sequences=True))
    model.add(tensorflow.keras.layers.LSTM(9))
    model.add(tensorflow.keras.layers.Dense(1))

    model.compile(optimizer = 'adam', loss = ['mse'])
    
    return model
```
