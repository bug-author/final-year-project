```python
def quantize_speeds(dataframe):
    """
    quantizes speeds for given dataframe to reduce computation

    returns: quantized dataframe
    """
    output_dataframe = pd.DataFrame()
    
    cols = dataframe.columns
    
    for col in cols:
        output_dataframe[col] = dataframe[col].round(decimals=3)
    
    return output_dataframe
        
```

```python
def get_positions(dataframe):
    """
    calculates positions for each vehicle for the trip

    returns: positions dataframe
    """
    output_dataframe = pd.DataFrame()
    
    cols = dataframe.columns
    
    for col in cols:
        output_dataframe[col] = dataframe[col].cumsum()
    
    return output_dataframe
        
```

```python
from pprint import pprint

pos_timestep_current = []

# best practices: dont use global variables


def cluster(dataframe, duration, cluster_radius, test_car = 'vehicle_1'):
    """
    clustering algorithm for offloading decision making

    returns: optimal cluster formation for given test vehicle
    """
    pos_cluster = {test_car: []}
    
    position_test = dataframe[test_car][duration] #test vehicle position for ending timestep
    
    all_pos_timestep = dataframe.iloc[duration, 1:] #positions data for all other cars for ending timestep
    

    for service_veh in range(len(all_pos_timestep)):
        diff = abs(all_pos_timestep[service_veh] - position_test)
        if diff < cluster_radius:
            pos_cluster[test_car].append({service_veh: round(diff, 2)})
    
    return pos_cluster

pprint(cluster(fab_data, duration=10, cluster_radius=5, test_car='vehicle_1'))
```

```python
class Cluster:
    
    def quantize_speeds(self, dataframe):
        """
        quantizes speeds for given dataframe to reduce computation

        returns: quantized dataframe
        """
        output_dataframe = pd.DataFrame()

        cols = dataframe.columns

        for col in cols:
            output_dataframe[col] = dataframe[col].round(decimals=3)

        return output_dataframe
   
    def get_positions(self, dataframe):
        """
        calculates positions for each vehicle for the trip

        returns: positions dataframe
        """
        output_dataframe = pd.DataFrame()

        cols = dataframe.columns

        for col in cols:
            output_dataframe[col] = dataframe[col].cumsum()

        return output_dataframe


    def get_clusters(self, dataframe, duration, cluster_radius, test_car = 'vehicle_1'):
        """
        clustering algorithm for offloading decision making

        returns: optimal cluster formation for given test vehicle
        """
        pos_cluster = {test_car: []}

        position_test = dataframe[test_car][duration] #test vehicle position for ending timestep

        all_pos_timestep = dataframe.iloc[duration, 1:] #positions data for all other cars for ending timestep


        for service_veh in range(len(all_pos_timestep)):
            diff = abs(all_pos_timestep[service_veh] - position_test)
            if diff < cluster_radius:
                pos_cluster[test_car].append({service_veh: round(diff, 2)})

        return pos_cluster

    
    
cluster_obj = Cluster()

quantized_df = cluster_obj.quantize_speeds(df)

positions_df = cluster_obj.get_positions(quantized_df)

clusters = cluster_obj.get_clusters(positions_df, duration=10, cluster_radius=5, test_car='vehicle_1')
```
