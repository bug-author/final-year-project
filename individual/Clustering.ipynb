```python
def quantize_speeds(dataframe):
    """
    quantizes speeds for given dataframe to reduce computation

    returns: quantized dataframe
    """
    output_dataframe = pd.DataFrame()
    
    cols = dataframe.columns
    
    for col in cols:
        output_dataframe[col] = dataframe[col].round(decimals=3)
    
    return output_dataframe
        
```

```python
def get_positions(dataframe):
    """
    calculates positions for each vehicle for the trip

    returns: positions dataframe
    """
    output_dataframe = pd.DataFrame()
    
    cols = dataframe.columns
    
    for col in cols:
        output_dataframe[col] = dataframe[col].cumsum()
    
    return output_dataframe
        
```
