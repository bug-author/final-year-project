```python
import os
import pandas as pd
```

```python
WORKING_DIR = os.getcwd()
BASE_DIR = os.path.dirname(WORKING_DIR)
DATA_DIR = os.path.join(BASE_DIR, 'data')
```

```python
os.listdir(DATA_DIR)
```

```python
dfs = []

for file in os.listdir(DATA_DIR):
    file_path = os.path.join(DATA_DIR, file)
    _ = pd.read_csv(file_path)
    dfs.append(_)
```

```python
df = pd.concat(dfs)
```

```python
df
```

```python
assert df.shape[0] >= 1471, "Shape mismatch in #rows"
assert df.shape[1] == 35, "Shape mismatch in #columns"
```
