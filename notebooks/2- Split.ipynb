```python
import os
import pandas as pd
```

```python
WORKING_DIR = os.getcwd()
BASE_DIR = os.path.dirname(WORKING_DIR)
DATA_DIR = os.path.join(BASE_DIR, 'data')
RAW_DATA_DIR = os.path.join(DATA_DIR, 'raw')
CACHE = os.path.join(DATA_DIR, 'cache')
```

```python
dfs = []

for file in os.listdir(RAW_DATA_DIR):
    file_path = os.path.join(RAW_DATA_DIR, file)
    _ = pd.read_csv(file_path, index_col='timestep_time')
    dfs.append(_)
```

```python
df = pd.concat(dfs)
```

```python
df
```

```python
assert df.shape[0] >= 1471, "Shape mismatch in #rows"
assert df.shape[1] == 34, "Shape mismatch in #columns"
```

```python
positions = df.iloc[:, :17]
speeds = df.iloc[:, 17:]
```

```python
assert positions.shape[1] == 17, "Invalid splitting"
assert speeds.shape[1] == 17, "Invalid splitting"
```

# Export

```python
positions_dataset = os.path.join(CACHE, 'positions.csv')
speeds_dataset = os.path.join(CACHE, 'speeds.csv')
```

```python
positions.to_csv(positions_dataset, index=False)
speeds.to_csv(speeds_dataset, index=False)
```
