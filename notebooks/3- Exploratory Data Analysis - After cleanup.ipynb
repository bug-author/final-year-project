```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import random

%matplotlib inline
plt.style.use('fivethirtyeight')


from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 18,10
```

```python
WORKING_DIR = os.getcwd()
BASE_DIR = os.path.dirname(WORKING_DIR)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE = os.path.join(DATA_DIR, 'cache')
```

# Positions

```python
positions_df = pd.read_csv(f'{CACHE}/positions_out.csv')
```

```python
speeds_df = pd.read_csv(f'{CACHE}/speeds_out.csv')
```

```python
plt.plot(positions_df);
plt.title('Vehicle Positions vs Time');
plt.xlabel('Time');
plt.ylabel('Position');
plt.legend(positions_df.columns);
```

## Observations
- Purely linear data
- Speculation: some vehicles might have correlations in positions (also observed while simulating data in SUMO)
- Data ranges from 0 to variable positions - Standardization/ Normalization is needed

# Speeds

```python
plt.plot(speeds_df);
plt.title('Vehicle Speeds vs Time');
plt.xlabel('Time');
plt.ylabel('Speed');
plt.legend(speeds_df.columns);
```

```python
rcParams['figure.figsize'] = 24,10
values = speeds_df.to_numpy()

for i, feature in enumerate(speeds_df.columns[0:5]):
  clr = ['#30a2da', '#fc4f30', '#e5ae38', '#6d904f', '#8b8b8b'];
  plt.subplot(len(speeds_df.columns[0:5]), 1, i+1);
  plt.plot(values[:100, i], color=clr[i%4], label=feature);
  plt.legend(loc='upper left')
plt.show();
```

```python
for i, feature in enumerate(speeds_df.columns[5:10]):
  clr = ['#30a2da', '#fc4f30', '#e5ae38', '#6d904f', '#8b8b8b'];
  plt.subplot(len(speeds_df.columns[5:10]), 1, i+1);
  plt.plot(values[:100, i], color=clr[i%4], label=feature);
  plt.legend(loc='upper left')
plt.show();
```
