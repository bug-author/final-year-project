```python
import pandas as pd
import os
```

```python
WORKING_DIR = os.getcwd()
BASE_DIR = os.path.dirname(WORKING_DIR)
DATA_DIR = os.path.join(BASE_DIR, 'data')
RAW_DATA_DIR = os.path.join(DATA_DIR, 'raw')
CACHE = os.path.join(DATA_DIR, 'cache')
```

```python
os.listdir(CACHE)
```

```python
train_speed = pd.read_csv(f'{CACHE}/speeds_train_out.csv')
test_speed = pd.read_csv(f'{CACHE}/speeds_test_out.csv')
```

```python
train_speed.shape
```

```python
test_speed.shape
```

```python
speed_final = pd.concat([train_speed, test_speed])
```

```python
cols = [f'Vehicle {x+1} Speed' for x in range(17)]
```

```python
speed_final.columns = cols
```

```python
speed_final
```

```python
train_position = pd.read_csv(f'{CACHE}/positions_train_out.csv')
test_position = pd.read_csv(f'{CACHE}/positions_test_out.csv')
```

```python
position_final = pd.concat([train_position, test_position])
```

```python
cols = [f'Vehicle {x+1} Position' for x in range(17)]
```

```python
position_final.columns = cols
```

```python
print(speed_final.shape)
print(position_final.shape)
```

```python
whole_dataset = pd.concat([speed_final, position_final], axis=1)
```

```python
whole_dataset = os.path.join(CACHE, 'whole_data_out.csv')
whole_dataset.to_csv(whole_dataset, index=False)
```
