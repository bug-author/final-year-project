```python
import pandas as pd
import requests
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import keras
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
```

```python
res = requests.get("https://ai-emp-fog-comp.herokuapp.com/raw-data")
response = res.json()
```

```python
df = pd.DataFrame.from_dict(response["raw_data"])
```

```python
df.drop(columns="timestep_time", inplace = True)
```

#Creating Dataset for feeding the model

```python
train_df = df.iloc[:1200, :]
test_df = df.iloc[1200:, :]
```

```python
df.head()
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
transformer = Normalizer()
transformer.fit(train_df)

train_df_scaled = transformer.transform(train_df)
test_df_scaled = transformer.transform(test_df)
```

```python
def create_dataset(data, steps):
  features, labels = [], []
  for i in range(len(data)-steps):
    data_seq = data[i: i+steps, :]
    features.append(data_seq)

    seq_label = data[i+steps, :]
    labels.append(seq_label)

  return np.array(features), np.array(labels)
```

```python
TIME_STEPS = 2

train_features, train_labels = create_dataset(train_df_scaled, TIME_STEPS)
test_features, test_labels = create_dataset(test_df_scaled, TIME_STEPS)
```

#2-Layer LSTM Model

```python
from keras.backend import dropout
model = keras.models.Sequential()
model.add(keras.layers.LSTM(50, input_shape=(train_features.shape[1], train_features.shape[2]), return_sequences=True) )
model.add(keras.layers.LSTM(5))
model.add(keras.layers.Dense(34))

model.compile(optimizer='adam', loss=['mse', 'mae'])

history = model.fit(train_features, train_labels, batch_size=64, epochs=120, shuffle=False, validation_split=0.1)

plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```

```python
min(history.history['val_loss'])
```

```python
output = model.predict(test_features)
# plt.plot(output)
```

```python
%matplotlib inline

plt.plot(test_df_scaled[0, 17:], label='Actual Speed');
plt.plot(output[0, 17:], label='Predicted Speed');
plt.legend()
```

```python
plt.plot(test_df_scaled[0, :17], label='Actual Position');
plt.plot(output[0, :17], label='Predicted Position');
plt.legend()
```

#3 Layer LSTM Model

```python
model = keras.models.Sequential()
model.add(keras.layers.LSTM(50, input_shape=(train_features.shape[1], train_features.shape[2]), return_sequences=True))
model.add(keras.layers.LSTM(32, return_sequences = True))
model.add(keras.layers.LSTM(8))
model.add(keras.layers.Dense(34))

model.compile(optimizer='adam', loss=['mse', 'mae'])
```

```python
model.summary()
```

```python
history = model.fit(train_features, train_labels, batch_size=64, epochs=100, validation_split=0.1)

plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```
